## Data manipulation


```python
# Parsing parse_date parameter to change the saledate column from an object to a datetime
df = pd.read_csv('data-files/TrainAndValid.csv', low_memory=False, parse_dates=["saledate"])
df.saledate[:5]
```

```python
df.sort_values(by=["saledate"], inplace=True, ascending=True)
```

```python
# A way of seeing all the columns
df.head(5).T
```

```python
# Creating a copy of the original df
df_tmp = df.copy()
```

```python
# Enriching the data frame
df_tmp["saleYear"] = df_tmp.saledate.dt.year
df_tmp["saleMonth"] = df_tmp.saledate.dt.month
df_tmp["saleDay"] = df_tmp.saledate.dt.day
df_tmp["saleDayOfWeek"] = df_tmp.saledate.dt.dayofweek
df_tmp["saleDayOfYear"] = df_tmp.saledate.dt.dayofyear

# Removing the saledate column
df_tmp.drop("saledate", axis =1, inplace=True)
```

### Converting strings/objects to categories

```python
pd.api.types.is_string_dtype(df_tmp["UsageBand"])
```

```python
for label, content in df_tmp.items():
    if pd.api.types.is_object_dtype(content):
        print(label)
```

```python
# Converting string to categories
for label, content in df_tmp.items():
    if pd.api.types.is_string_dtype(content) or pd.api.types.is_object_dtype(content):
        df_tmp[label] = content.astype("category").cat.as_ordered()
```

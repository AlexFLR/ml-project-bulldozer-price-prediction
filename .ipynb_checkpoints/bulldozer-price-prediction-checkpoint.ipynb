## Importing libraries/modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_log_error, mean_absolute_error, r2_score
```

```python
# Creating a copy of the original df
df_tmp = df.copy()
```

```python
# Enriching the data frame
df_tmp["saleYear"] = df_tmp.saledate.dt.year
df_tmp["saleMonth"] = df_tmp.saledate.dt.month
df_tmp["saleDay"] = df_tmp.saledate.dt.day
df_tmp["saleDayOfWeek"] = df_tmp.saledate.dt.dayofweek
df_tmp["saleDayOfYear"] = df_tmp.saledate.dt.dayofyear

# Removing the saledate column
df_tmp.drop("saledate", axis =1, inplace=True)
```

## Fill missing values

```python
for label, content in df_tmp.items():
    if pd.api.types.is_numeric_dtype(content):
        print(label)
```

```python
# Check what numeric columns have null values:
for label, content in df_tmp.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
        
```

```python
#Fill the missing values with the  median ( more robust than mean)
for label,content in df_tmp.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            df_tmp[label+'_is_missing'] = pd.isnull(content)
            df_tmp[label] = content.fillna(content.median())
            
```

```python
for label,content in df_tmp.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
```

### Filling categorical variables into numbers and fill missing values


```python
for label, content in df_tmp.items():
    if not pd.api.types.is_numeric_dtype(content):
        print(label)
```

```python
for label, content in df_tmp.items():
    if not pd.api.types.is_numeric_dtype(content):
        df_tmp[label+'is_missing'] = pd.isnull(content)
        df_tmp[label] = pd.Categorical(content).codes + 1 # for missing values it assigns -1  and we dont need neg. values
```

### Building the machine learning model

```python
# Splitting the set into train and validation 
# validation set -> only from 2012
df_val = df_tmp[df_tmp.saleYear == 2012]
df_train = df_tmp[df_tmp.saleYear != 2012]
len(df_val), len(df_train)
```

```python
X_train, y_train = df_train.drop('SalePrice', axis=1), df_train.SalePrice
X_valid, y_valid = df_val.drop('SalePrice', axis =1), df_val.SalePrice
X_train.shape, y_train.shape, X_valid.shape, y_valid.shape
```

```python
# Build an evaluation function for RMSLE


def rmsle(y_test, y_preds):
    return np.sqrt(mean_squared_log_error(y_test, y_preds))

# Create function to evaluate model on a few different levels
def show_score(model):
    train_preds = model.predict(X_train)
    valid_preds = model.predict(X_valid)
    scores = {"Training MAE": mean_absolute_error(y_train, train_preds),
             "Valid MAE":mean_absolute_error(y_valid, valid_preds),
             "Training RMSLE":rmsle(y_train, train_preds),
             "Valid RMSLE":rmsle(y_valid, valid_preds),
             "Training R^2":r2_score(y_train,train_preds),
             "Valid R^2":r2_score(y_valid, valid_preds)}
    return scores
    
```

# Testing the model on a subset (used to tune the hyperparameters)

```python
model = RandomForestRegressor(n_jobs =-1,
                             random_state=42,
                             max_samples=10000)
```

```python
%%time
#Cutting down on the max number samples  each estimator can see improves training time
model.fit(X_train,y_train)
```

```python
show_score(model)
```

## Hyperparameter Tuning ->RandomizedSearchCV

```python
model = RandomForestRegressor()
print(model.get_params())
```

```python
%%time
rf_grid = {
'n_estimators': np.arange(10,100,10),
    'max_depth':[None,3,5,10],
    'min_samples_split':np.arange(2,20,2),
    'min_samples_leaf':np.arange(1,20,2),
    'max_features':[0.5,1,'sqrt','auto'],
    'max_samples':[10000]
}
rs_model = RandomizedSearchCV(RandomForestRegressor(n_jobs = -1,
                                                     random_state=42),
                              param_distributions=rf_grid,
                              n_iter = 5,
                              cv = 5,
                              verbose = True)
# Fitting the model
rs_model.fit(X_train, y_train)
```

```python
# Training a model with 100 iter.
rs_model2 = RandomizedSearchCV(RandomForestRegressor(n_jobs = -1,
                                                     random_state=42),
                              param_distributions=rf_grid,
                              n_iter = 100,
                              cv = 5,
                              verbose = True)
```

```python
rs_model.best_params_, rs_model2.best_params_
```

## Testing the trained model

```python
show_score(rs_model),show_score(rs_model2)
```
